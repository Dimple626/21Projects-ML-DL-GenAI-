```python
# Step 1: Install and import libraries
!pip install ydata-profiling

import pandas as pd
import numpy as np

# Step 2: Load the dataset (from the directory as in your notebook)
# If already cloned:
data_path = '21-Days-21-Projects-Dataset/Datasets/Titanic-Dataset.csv'
titanicdf = pd.read_csv(data_path)

# Step 3: Data Cleaning
# Fill missing Age with median
titanicdf['Age'] = titanicdf['Age'].fillna(titanicdf['Age'].median())

# Fill missing Embarked with mode
titanicdf['Embarked'] = titanicdf['Embarked'].fillna(titanicdf['Embarked'].mode()[0])

# Create a HasCabin feature instead of Cabin (drop Cabin)
titanicdf['HasCabin'] = titanicdf['Cabin'].notna().astype(int)
titanicdf.drop('Cabin', axis=1, inplace=True)

# Step 4: Feature engineering (optional, recommended by your notebook)
# FamilySize: number of siblings/spouses plus parents/children plus self
titanicdf['FamilySize'] = titanicdf['SibSp'] + titanicdf['Parch'] + 1
# IsAlone: 1 if alone, 0 otherwise
titanicdf['IsAlone'] = (titanicdf['FamilySize'] == 1).astype(int)

# Step 5: Biprofiler (ydata-profiling)
from ydata_profiling import ProfileReport
profile = ProfileReport(titanicdf, title="Titanic Biprofiler Report", explorative=True)

# Save the report as HTML file
profile.to_file("Titanic_Biprofiling_Report.html")

# To display in the notebook (optional, for interactive use)
# profile.to_widgets()

# If you want to see the cleaned data preview
print(titanicdf.head())
```
